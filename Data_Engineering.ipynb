```python
import pandas as pd
import numpy as np
import os
import datetime
```

```python
# CSV file
csvfile = "Resources/hawaii_measurements.csv"
csvfile_2 = "Resources/hawaii_stations.csv"
```

```python
dataframe_HM = pd.read_csv(csvfile,dtype = object)
dataframe_HM.dropna(axis = 0)
```

```python
#datetime.strptime(dataframe_HM['date'],'%Y-%m-%d')
date = [i for i in dataframe_HM['date']]
dates = []
for i in date:
    dates.append((datetime.datetime.strptime(i,'%Y-%m-%d')).date())
```

```python
dataframe_HM.drop(['date'], axis = 1, inplace = True)
```

```python
dataframe_HM['date'] = dates
```

```python
dataframe_HS = pd.read_csv(csvfile_2,dtype = object)
dataframe_HS.dropna(axis = 0)
```

```python
clean_data_1 = "cleaned_Hawaii_Measurements.csv"
clean_data_2 = "cleaned_Hawaii_Stations.csv"
dataframe_HM.to_csv(clean_data_1, index=False)
dataframe_HS.to_csv(clean_data_2, index=False)
```
